# Getting data using BS4


```python
#Importing all needed libraries
import pandas as pd
import time
import requests
from pprint import pprint
from bs4 import BeautifulSoup
from pprint import pprint

import json
from lxml import html
import re
import csv
```

```python
#Getting the url and making it a BS4 object
url="https://www.youtube.com/feed/trending"
response=requests.get(url)
page=response.text #(kam content)
page=BeautifulSoup(page, "html.parser")

#outlining the div with all the necessary info on titles, urls, duration etc

trend_div_all=page.find_all("div", class_="yt-lockup-content")                                      
trend_div_all[0]
```

```python
#getting trending video urls
urls_inc=[i.find("a", attrs={"dir":"ltr"}).get("href") for i in trend_div_all]
urls=["https://www.youtube.com"+str(i)for i in urls_inc]
urls
```

```python
#getting titles
titles=[i.find("a", attrs={"dir":"ltr"}).get("title") for i in trend_div_all]
titles
```

```python
#getting duration and cleaning it

dur_tag=[i.find_all("span", class_="accessible-description") for i in trend_div_all]
durations=[i[0].get_text() for i in dur_tag]
duration_cl=[i.replace("- Duration: ","") for i in durations]
duration= [i.replace(".","") for i in duration_cl]
duration
```

```python
#getting username

username=[i.find("a", class_="yt-uix-sessionlink spf-link ").get_text() for i in trend_div_all]
username
```

```python
#getting views and cleaning data


view=[i.get_text() for i in page.find_all("li", text=re.compile("views"))]
views=[i.replace("views", "")for i in view]
views
```

```python
#making the obtained data a pandas dataframe to sum up
final_bs4=pd.DataFrame(list(zip(titles,urls,duration, username, views)), columns=['Title','Url', 'Duration', 'Username', 'Views'])

final_bs4 
```

```python
#Saving to csv with no separate indexing and using colnames as headers
final_bs4.to_csv('final_bs4.csv',index=False, header=True)
#opening csv to verify everything is saved correctly
data=pd.read_csv('final_bs4.csv')
pprint(data.head(5))
```

```python
#Saving to json file with orient=records to match colnames and values

final_bs4.to_json('final_bs4.json', orient="records") 
#opening json to read it
with open('final_bs4.json','r') as final_bs4:
    bs4_json = json.load(final_bs4)
pprint(bs4_json)
```
